#### Changing the 'gender' column format
It takes out the capital letter at the beginning and the \n at the end

```python
import pandas as pd
import os
folder = '/home/maria/Escriptori/I2RCED/i2rced/organisations/complete_organisations'
for file_name in os.listdir(folder):
        if file_name.endswith('.csv'):  # Process only CSV files
            file_path = os.path.join(folder, file_name)

            # Read the CSV file
            df = pd.read_csv(file_path)
            df['gender'] = df['gender'].str.replace('\n', '').str.lower()
            df.to_csv(file_path, index=False)
```

#### Summary of all organisations
It takes the 192 organisations and counts how many users there are and how many women. The information is stored in *organisations_summary* with the columns **organisation**, **num_users** and **num_females**.

```python
folder = '/home/maria/Escriptori/I2RCED/i2rced/organisations/complete_organisations'
output_file = 'organisations_summary.csv'

organisation_summary = []
for file_name in os.listdir(folder):
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder, file_name)
            
            df = pd.read_csv(file_path)
          
            name = file_name.split('_')[1]
        
        # Count total rows and rows where 'female' is present
            total_rows = len(df)
            num_females = df[df.isin(['female']).any(axis=1)].shape[0]
        
        # Add to the summary data
            organisation_summary.append({'organisation': name, 'num_users': total_rows, 'num_females': num_females})

summary_df = pd.DataFrame(organisation_summary)
summary_df.to_csv(output_file, index=False)
```

```python
summary = pd.read_csv('organisations_summary.csv')
summary['% of women'] = (summary['num_females']/summary['num_users'])*100
summary['% of women'] = summary['% of women'].round(2)
summary.to_csv('organisations_summary.csv', index=False)
```
